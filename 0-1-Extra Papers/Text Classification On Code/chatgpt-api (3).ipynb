```python
import pandas as pd
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
```

```python
test_df=pd.read_csv("")#datapath
```

```python
test_df
```

```python
def chat_with_chatgpt(prompt, model="gpt-3.5-turbo"):
    chat_completion=client.chat.completions.create(
        messages=[
            {"role":"user",
             "content":prompt
            }
        ],
        model="gpt-3.5-turbo"
    )
    message = chat_completion.choices[0].message.content
    return message
```

```python
from tqdm import tqdm

text = []

for index, row in tqdm(enumerate(test_df['text']), total=len(test_df)):
    user_prompt = '''For the given Input  "{}", is the hate in the in input Geopolitical, Neutral,  Personal, Political, and  Religious ? write only your response nothing else. Don't add anything before and after your response.'''.format(row) 
    chatbot_response = chat_with_chatgpt(user_prompt)
    text.append(chatbot_response)

    if index % 500 == 0:
        print(index)
```

```python
df2 = pd.DataFrame({'label': text})

df2.to_csv('predictions.csv', index=False)  
```

```python
df2.head()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report

predicted_labels = df2['label'].to_list()
true_labels = test_df['label'].to_list()  
label_encoder = LabelEncoder()

true_labels_encoded = label_encoder.fit_transform(true_labels)


predicted_labels_encoded = label_encoder.fit_transform(predicted_labels)

print(classification_report(true_labels_encoded, predicted_labels_encoded,digits=4))
```
